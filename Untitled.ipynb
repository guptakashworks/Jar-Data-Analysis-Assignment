```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("Walmart Sales.xlsx")
```

<H1>Data Analysis</H1>

```python
Product_Seg=list(df["Product line"].unique())
```

```python
Product_Seg
```

```python
df.columns
```

```python
# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract month from the 'Date' column
df['Month'] = df['Date'].dt.month

# Calculate total sales for each transaction
df['Total Sales'] = df['Quantity'] * df['Unit price']

data =df.copy()
```

```python
data
```

```python




# Group data by month and calculate total sales
monthly_sales = data.groupby('Month')['Total Sales'].sum()

# Calculate the average monthly sales
average_monthly_sales = monthly_sales.mean()

# Plot the total monthly sales and average monthly sales
plt.figure(figsize=(7,6))

bars = plt.bar(monthly_sales.index, monthly_sales.values, alpha=0.7, label='Total Monthly Sales')


for bar, sales in zip(bars, monthly_sales.values):
    if sales < average_monthly_sales:
        bar.set_color('red')
    else:
        bar.set_color('green')
        
plt.axhline(y=average_monthly_sales, color='red', linestyle='--', label='Average Monthly Sales')
plt.text(0.75,average_monthly_sales*0.95, f'Average Monthly Sales: ${average_monthly_sales:.2f}', color='red', ha='left',va='center')
plt.title('Total Monthly Sales with Average Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.legend()
plt.xticks(rotation=0)  # Rotate x-axis labels if necessary
plt.grid(axis='y', linestyle='--', alpha=0.7)

for bar, sales in zip(bars, monthly_sales.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height()/2, f'${sales:.2f}', ha='center', va='bottom',rotation=90)
plt.show()
```

<H1> Sales based on city and Branch Level</H1>

```python
city_sales_df = data.groupby(["City", "Branch"], as_index=False)["Total Sales"].sum().copy()

city_sales_df
```

```python


# Plot the bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(city_sales_df['City'] + ' - ' + city_sales_df['Branch'], city_sales_df['Total Sales'], color='skyblue', alpha=0.7)
average_city_sales=city_sales_df["Total Sales"].mean()
# Annotate exact values on top of each bar
for bar, sales in zip(bars, city_sales_df['Total Sales']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height()/2, f'${sales:.2f}', ha='center', va='bottom', rotation=90)

plt.axhline(y=average_city_sales, color='red', linestyle='--', label='Average City Sales')
plt.text(0.75,average_city_sales*0.95, f'Average City Sales: ${average_city_sales:.2f}', color='red', ha='left',va='center')
plt.title('Total Sales by City & Branches')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels due to over lapping 
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

<H1> Average Price of item Sold at City and Branch Level</H1>

```python
city_avg_price_perUnit = data.groupby(['City', 'Branch']).agg({'Total Sales': 'sum', 'Quantity': 'sum'}).reset_index()
```

```python
city_avg_price_perUnit["Average Price"] = city_avg_price_perUnit["Total Sales"]/city_avg_price_perUnit["Quantity"]
city_avg_price_perUnit
```

```python
plt.bar(city_avg_price_perUnit["City"]+ " - " + city_avg_price_perUnit["Branch"], city_avg_price_perUnit["Average Price"], color='skyblue', alpha=0.7)

plt.xlabel('City - Branch')
plt.ylabel('Total Sales')
plt.title('Total Sales by City and Branch')

plt.xticks(rotation=45)  # Rotate x-axis labels due to over lapping
plt.tight_layout()
```

```python
grouped_data = df.groupby(['City', 'Branch'])

for name, group in grouped_data:
    city, branch = name
    plt.figure(figsize=(14, 12))
    
    # Heading for Transaction charts
    plt.suptitle(f'{city} - Branch {branch} - Transaction vs Total Sales', fontsize=16)
    
    # First row - Number of transactions
    plt.subplot(2, 3, 1)
    gender_counts = group['Gender'].value_counts()
    plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title('Gender Distribution (Transactions)')
    plt.axis('equal')
    
    plt.subplot(2, 3, 2)
    product_line_counts = group['Product line'].value_counts()
    plt.pie(product_line_counts, labels=product_line_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title('Product Line Distribution (Transactions)')
    plt.axis('equal')
    
    plt.subplot(2, 3, 3)
    payment_method_counts = group['Payment'].value_counts()
    plt.pie(payment_method_counts, labels=payment_method_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title('Payment Method Distribution (Transactions)')
    plt.axis('equal')
    
    # Second row - Total sales
    plt.subplot(2, 3, 4)
    gender_sales = group.groupby('Gender')['Total Sales'].sum()
    plt.pie(gender_sales, labels=gender_sales.index, autopct='%1.1f%%', startangle=90)
    plt.title('Gender Distribution (Total Sales)')
    plt.axis('equal')
    
    plt.subplot(2, 3, 5)
    product_line_sales = group.groupby('Product line')['Total Sales'].sum()
    plt.pie(product_line_sales, labels=product_line_sales.index, autopct='%1.1f%%', startangle=90)
    plt.title('Product Line Distribution (Total Sales)')
    plt.axis('equal')
    
    plt.subplot(2, 3, 6)
    payment_method_sales = group.groupby('Payment')['Total Sales'].sum()
    plt.pie(payment_method_sales, labels=payment_method_sales.index, autopct='%1.1f%%', startangle=90)
    plt.title('Payment Method Distribution (Total Sales)')
    plt.axis('equal')
    
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust the layout to include the suptitle
    plt.show()
```
